#                                           Rock vs Mine Prediction

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
df=pd.read_csv('/Users/vaishalisingh/Desktop/sonar_data.csv')
df.head()
```

```python
df.describe()
```

```python
x=df.drop(df.columns[60], axis='columns')
x.head()
```

```python
y=df.R
y
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

## Model Selection

```python
model_1=cross_val_score(LogisticRegression(random_state=2),x,y)
print(model_1)
print(np.mean(model_1))
```

```python
model_2=cross_val_score(SVC(kernel="linear",random_state=2),x,y)
print(model_2)
print(np.mean(model_2))
```

```python
model_3=cross_val_score(tree.DecisionTreeClassifier(splitter="random",random_state=2),x,y)
print(model_3)
print(np.mean(model_3))
```

```python
model_4=cross_val_score(RandomForestClassifier(random_state=2),x,y)
print(model_4)
print(np.mean(model_4))
```

### Random Forest had the better score 
## Applying random forest on the dataset:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
model=RandomForestClassifier(random_state=2)
model.fit(x_train,y_train)
```

## Model Evaluation

```python
x_train_prediction=model.predict(x_train)
training_accuracy=accuracy_score(x_train_prediction,y_train)
print('Accuracy on training data: ',training_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
testing_accuracy=accuracy_score(x_test_prediction,y_test)
print('Accuracy on testing data: ',testing_accuracy)
```

```python
print('Model Score: ',model.score(x_test,y_test))
```
